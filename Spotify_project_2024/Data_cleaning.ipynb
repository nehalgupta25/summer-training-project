```python
#importing libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
#loading csv data file
import pandas as pd 
df=pd.read_csv('Most Streamed Spotify Songs 2024.csv')
df.head(20)
```

```python
#generating information of the data
df.info()
```

```python
#renaming specified columns
df.rename(columns={
    'Spotify Streams':'Streams',
    'Spotify Playlist Count':'Playlist_count',
    'Spotify Playlist Reach':'Playlist_reach',
    'Spotify Popularity':'Popularity'
    },inplace =True
)
print("columns rename successfully.")
```

```python
#checking and counting null values 
pd.isnull(df).sum()
```

```python
#showing all the columns
print(df.columns)
```

```python
#filling missing values
missing_value = {
    'Artist',
    'Streams', 
    'Playlist_count',
    'Playlist_reach',
    'Popularity'
}

for col in missing_value:
    df[col] = df[col].fillna(0)

print("Missing values successfully replaced.")
```

```python
#updating data
df['Artist']=df['Artist'].replace(0,'Unknown')
```

```python
#checkinh total null values
pd.isnull(df).sum()
```

```python

df.duplicated().sum()
```

```python
#droping all duplicated data
df.drop_duplicates(inplace=True)
print("duplicates values deleted.")
```

```python
#checking duplicates again
df.duplicated().sum()
```

```python
#filtering data

df[df['Track Score']>100]
df[df['Explicit Track'] == True]
```

```python
unknown_album_count = df[df['Album Name'] == 'Unknown']
print(f"Tracks with Unknown Album: {unknown_album_count}")
```

```python
#checking unique data
print("Rows and Columns:",df.shape)
print("Unique Artist:")
print(df['Artist'].unique())
print("Unique data:")
df.nunique()
```

```python
#changing data types of columns
df['Streams'] = df['Streams'].astype(str).str.replace(',', '', regex=False)
df['Streams']=pd.to_numeric(df['Streams'])

df['Playlist_count'] = df['Playlist_count'].astype(str).str.replace(',', '', regex=False)
df['Playlist_count']=pd.to_numeric(df['Playlist_count'])

df['Playlist_reach'] = df['Playlist_reach'].astype(str).str.replace(',', '', regex=False)
df['Playlist_reach']=pd.to_numeric(df['Playlist_reach'])

df['Release Date'] = pd.to_datetime(df['Release Date'], format='%d-%m-%Y', errors='coerce')

```

```python
#filtering data
df[df['Artist']== 'Unknown']
```

```python
#mean,median and stat of data based on popularity column
Popularity=df['Popularity']

print("Mean: ",np.mean(Popularity))
print("Median: ",np.median(Popularity))
print("Std Deviation: ",np.std(Popularity))
```

```python
df['Release Date']=df['Release Date'].replace(np.nan,'Unknown')
```

```python
#cleaned data file
df.to_csv("Spotify_Cleaned_Data1.csv",index=False)
print("cleaned sspotify file created.")
```

```python
# Performing Exploratory Data Analysis (EDA)
 # visualizations to show insights:
#  With Seaborn / Matplotlib:
```

```python
top_tracks = df.sort_values('Streams', ascending=False).head(10)
#top 10 songs
plt.figure(figsize=(12, 6))
sns.barplot(x='Streams', y='Track', data=top_tracks, hue='Track', palette="viridis")
plt.title('Top 10 Tracks by Spotify Streams')
plt.xlabel('Streams')
plt.ylabel('Track')
plt.tight_layout()
plt.show()

#stream vs popularity(scatter)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Popularity', y='Streams', hue='Explicit Track', data=df)
plt.title('Streams vs Popularity')
plt.xlabel('Popularity')
plt.ylabel('Streams')
plt.tight_layout()
plt.show()

#top 10 artist
top_artists = df['Artist'].value_counts().head(10).index
filtered_df = df[df['Artist'].isin(top_artists)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='Artist', y='Streams', data=filtered_df)
plt.xticks(rotation=45)
plt.title('Streams Distribution by Top 10 Artists')
plt.tight_layout()
plt.show()


#Playlist Count vs Playlist Reach
plt.figure(figsize=(8, 6))
sns.histplot(x='Playlist_count', y='Playlist_reach', data=df,color='skyblue')
plt.title('Playlist Count vs Playlist Reach')
plt.xlabel('Playlist Count')
plt.ylabel('Playlist Reach')
plt.tight_layout()
plt.show()







```
